```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from Boosting import GradientBoostingReg
from sklearn.ensemble import GradientBoostingRegressor
```

```python
def square_error(y, y_hat):
    error = y - y_hat
    square = error.dot(error.T)
    return square

def r_square(ssr, sst):
    return ssr / sst
```

```python
x = np.random.randn(100)
y = 4*x**3 + 5*x + 4 + 2*np.random.randn(100)
```

```python
plt.scatter(x,y)
```

```python
x=np.array(x).reshape(-1,1)
y=np.array(y)
model=GradientBoostingReg(n_estimators=2000 , learning_rate=0.01)
model.fit(x,y)
y_hat=model.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,y_hat,"r.")
```

```python
sse = square_error(y,y_hat)
sst = square_error(y,y.mean())
ssr = sst - sse
r_square(ssr,sst)
```

```python
model=GradientBoostingRegressor()
model.fit(x,y)
y_hat=model.predict(x)
plt.scatter(x,y)
plt.plot(x,y_hat , "r.")
```

```python
sse = square_error(y,y_hat)
sst = square_error(y,y.mean())
ssr = sst - sse
```

```python
r_square(ssr,sst)
```
