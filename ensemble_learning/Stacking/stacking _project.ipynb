```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from Stacking_implem import Stacking
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
def square_error(y, y_hat):
    error = y - y_hat
    square = error.dot(error.T)
    return square

def r_square(ssr, sst):
    return ssr / sst
```

```python
x = np.random.randn(100)
y = 4*x**3 + 5*x**4 + 4*x + 2*np.random.randn(100)
```

```python
plt.scatter(x,y)
```

```python
x=np.array(x).reshape(-1,1)
y=np.array(y)
model=Stacking(base_models=[RandomForestRegressor() ], final_model=GradientBoostingRegressor())
model.fit(x,y)
y_hat=model.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,y_hat,"r.")
```

```python
sse = square_error(y,y_hat)
sst = square_error(y,y.mean())
ssr = sst - sse
r_square(ssr,sst)
```
